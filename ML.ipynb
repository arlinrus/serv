## Реальный вариант РК по ML
### + простейший способ решения

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('https://courses.openedu.ru/assets/courseware/v1/431a30c4a3a2b3f265f8cce1a809aa0f/asset-v1:ITMOUniversity+MLDATAN+spring_2023_ITMO_bac+type@asset+block/pulsar_stars_new.csv')

dfq = df.query('(TG == 0 & MIP >=  59.203125 & MIP <= 192.6171875) | (TG == 1 & MIP >= 5.8125 & MIP <= 115.6875)')
dfq.STDC.mean()
```

```python
dfs = dfq.sort_values(by='SIP')
X = dfs.drop('TG',axis=1)
Y = dfs['TG']

x,xt,y,yt = train_test_split(X,Y,random_state=25,test_size=0.2,stratify=Y)
x.EKC.mean()
```

```python
sc = MinMaxScaler().fit(x)
x[x.columns] = sc.transform(x)
xt[xt.columns] = sc.transform(xt)
x.MIP.mean()
```

```python
pred = LogisticRegression(random_state = 25).fit(x,y).predict(xt)
confusion_matrix(yt,pred)
```

```python
f1_score(yt,pred)
```

```python
pred = KNeighborsClassifier(n_neighbors=3).fit(x,y).predict(xt)
confusion_matrix(yt,pred)
```

```python
f1_score(yt,pred)
```
